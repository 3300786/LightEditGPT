```python
# Cell 1: 导入依赖
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from peft import PeftModel
import torch, json
import pandas as pd

# Cell 2: 加载模型
model_name = "gpt2"
adapter_path = "../output"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
base_model = AutoModelForCausalLM.from_pretrained(model_name)
model = PeftModel.from_pretrained(base_model, adapter_path)
model = model.to("cuda" if torch.cuda.is_available() else "cpu")
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, device=0 if torch.cuda.is_available() else -1)

# Cell 3: 加载数据
with open("../data/edit_dataset.jsonl", "r", encoding="utf-8") as f:
    lines = f.readlines()
    dataset = [json.loads(l) for l in lines]

# Cell 4: 生成预测并对比
records = []
for item in dataset:
    prompt = item["prompt"]
    target = item["target"].lower().strip()
    output = generator(prompt, max_new_tokens=30)[0]["generated_text"][len(prompt):].strip().lower()
    success = target.split()[0] in output
    records.append({
        "Prompt": prompt.strip(),
        "Expected": target,
        "Generated": output,
        "Success": success
    })

df = pd.DataFrame(records)
df.style.applymap(lambda v: 'background-color: lightgreen' if v is True else 'background-color: pink', subset=["Success"])
```

```python
# ✅ 漂移性分析（控制组）

import json

control_path = "../data/control_dataset.jsonl"

# 加载对照组样本
with open(control_path, "r", encoding="utf-8") as f:
    control_data = [json.loads(line) for line in f]

control_records = []
for item in control_data:
    prompt = item["prompt"]
    target = item["target"].lower().strip()
    output = generator(prompt, max_new_tokens=30)[0]["generated_text"][len(prompt):].strip().lower()
    success = target.split()[0] in output
    control_records.append({
        "Prompt": prompt.strip(),
        "Expected": target,
        "Generated": output,
        "Unchanged": success
    })

control_df = pd.DataFrame(control_records)
display(control_df.style.applymap(
    lambda v: 'background-color: lightgreen' if v is True else 'background-color: pink',
    subset=["Unchanged"]
))

correct_count = sum([r["Unchanged"] for r in control_records])
print(f"✅ Control Drift Evaluation: {correct_count}/{len(control_records)} correct | Drift Rate: {(1 - correct_count / len(control_records)):.2%}")
```

🔍 After injecting knowledge "Tesla CEO → Elon Musk", we evaluated the drift across 7 unrelated factual prompts. Only 1 out of 7 preserved the correct answer. The observed drift rate of 85.71% indicates that the current editing approach (LoRA over multiple layers) leads to substantial interference in unrelated knowledge, suggesting the need for more localized and constrained editing mechanisms.
